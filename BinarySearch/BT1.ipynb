# Tư Duy & Giải Thuật: Xây Dựng Cây Biểu Thức (Expression Tree)

**Bài toán:** Chuyển đổi một biểu thức toán học trung tố (Infix) thành cây nhị phân và duyệt cây theo 3 thứ tự.
**Ví dụ:** `A=3+5*2-7-(6+b)`

## 1. Chiến lược tổng quan (The Big Picture)

Việc dựng cây trực tiếp từ biểu thức Infix (`A + B`) rất phức tạp vì phải xử lý độ ưu tiên toán tử và các dấu ngoặc lồng nhau. Giải pháp tối ưu là đi đường vòng qua **Postfix (Hậu tố)**:

1.  **Bước 1 (Parsing):** Chuyển `Infix` $\rightarrow$ `Postfix` (Khử bỏ ngoặc, sắp xếp theo thứ tự thực thi).
2.  **Bước 2 (Building):** Dùng chuỗi `Postfix` dựng `Tree` (Dựng từ dưới lên).
3.  **Bước 3 (Traversal):** Dùng Đệ quy để duyệt cây.

## 2. Chi tiết từng giai đoạn

### Giai đoạn 1: Infix sang Postfix (Shunting-yard Algorithm)
**Mục tiêu:** Sắp xếp lại các phần tử sao cho toán tử nào cần tính trước thì đứng trước, toán tử tính sau đứng sau.

* **Công cụ:** 1 `Stack` (chứa toán tử) và 1 `List` (chứa kết quả).
* **Độ ưu tiên (Precedence):** `^` > `* /` > `+ -` > `=`

**Luật di chuyển:**
1.  **Gặp Toán hạng (Số/Chữ):** $\rightarrow$ Đưa ngay vào `List` kết quả.
2.  **Gặp Dấu `(`:** $\rightarrow$ Đẩy vào `Stack` (đánh dấu bắt đầu phạm vi ưu tiên).
3.  **Gặp Dấu `)`:** $\rightarrow$ Lấy (`pop`) toàn bộ toán tử trong `Stack` ra `List` cho đến khi gặp dấu `(`. (Xử lý xong ngoặc).
4.  **Gặp Toán tử (`op1`):**
    * So sánh với toán tử đang nằm trên đỉnh Stack (`op2`).
    * **Nếu độ ưu tiên `op2` $\ge$ `op1`:** `Pop` `op2` ra `List` (vì `op2` mạnh hơn hoặc bằng, nó phải được tính trước). Lặp lại cho đến khi `Stack` rỗng hoặc gặp `op2` yếu hơn.
    * Sau đó, đẩy `op1` vào `Stack`.

> **Kết quả ví dụ:** `A=3+5*2-7-(6+b)` $\rightarrow$ `A 3 5 2 * + 7 - 6 b + - =`

### Giai đoạn 2: Postfix sang Expression Tree
**Mục tiêu:** Biến chuỗi tuyến tính thành cấu trúc phân cấp.
**Tư duy:** Postfix có đặc điểm: *"Hai toán hạng đi trước, toán tử đi sau sẽ xử lý hai toán hạng đó"*.

* **Công cụ:** 1 `Stack` (Lưu trữ các **Node** của cây, không phải ký tự).

**Luật dựng cây:**
Duyệt chuỗi Postfix từ trái sang phải:
1.  **Nếu là Toán hạng:**
    * Tạo `Node(Toán hạng)`.
    * Đẩy `Node` vào `Stack`.
2.  **Nếu là Toán tử:**
    * Tạo `Node(Toán tử)`.
    * `Pop` `Stack` lần 1 $\rightarrow$ Gán vào **CON PHẢI** (`Right Child`).
    * `Pop` `Stack` lần 2 $\rightarrow$ Gán vào **CON TRÁI** (`Left Child`).
    * Đẩy `Node(Toán tử)` này (bây giờ đã là gốc của một cây con) ngược lại vào `Stack`.

> **⚠️ Lưu ý quan trọng:** Stack là LIFO (Vào sau ra trước). Phần tử lấy ra đầu tiên luôn là toán hạng đứng sau trong phép tính $\rightarrow$ Phải gán vào **CON PHẢI**. Nếu gán nhầm, `3 - 5` sẽ thành `5 - 3`.

### Giai đoạn 3: Duyệt cây (Tree Traversal)
Sử dụng đệ quy (Recursion) vì bản chất cây là cấu trúc đệ quy.

1.  **Tiền tự (Pre-order):** `Gốc` $\rightarrow$ `Trái` $\rightarrow$ `Phải`
    * *Kết quả:* Ra dạng Prefix (Ký pháp Ba Lan).
2.  **Trung tự (In-order):** `Trái` $\rightarrow$ `Gốc` $\rightarrow$ `Phải`
    * *Kết quả:* Khôi phục lại biểu thức Infix ban đầu.
3.  **Hậu tự (Post-order):** `Trái` $\rightarrow$ `Phải` $\rightarrow$ `Gốc`
    * *Kết quả:* Ra dạng Postfix.

## 3. Ví dụ minh họa (Trace)

**Input:** `A=3+5*2-7-(6+b)`

**Bước 1: Tạo Postfix**
`Stack` hoạt động như sau khi gặp toán tử:
* Gặp `*`: Đẩy vào (vì ưu tiên cao hơn `+`).
* Gặp `-`: Pop `+` ra (vì `-` và `+` ngang nhau, ai đến trước tính trước).
* Gặp `(`: Reset ưu tiên cục bộ.

$\rightarrow$ Output: `A352*+7-6b+-=`

**Bước 2: Dựng cây (Stack Node)**

| Ký tự | Hành động | Trạng thái Stack (Các Node) |
| :--- | :--- | :--- |
| `A`, `3`, `5`, `2` | Đẩy Node vào | `[A, 3, 5, 2]` |
| `*` | Pop 2, 5. Tạo cây `(5*2)` | `[A, 3, (5*2)]` |
| `+` | Pop (5*2), 3. Tạo cây `(3+(5*2))` | `[A, (3+...)]` |
| `7` | Đẩy Node vào | `[A, (3+...), 7]` |
| `-` | Pop 7, (3+...). Tạo cây trừ | `[A, ((3+...)-7)]` |
| `6`, `b` | Đẩy Node vào | `[A, Cluster1, 6, b]` |
| `+` | Pop b, 6. Tạo cây `(6+b)` | `[A, Cluster1, (6+b)]` |
| `-` | Pop (6+b), Cluster1. Tạo cây trừ | `[A, (Cluster1 - (6+b))]` |
| `=` | Pop Cluster2, A. Tạo cây gán | `[(A = Cluster2)]` |

$\rightarrow$ **Gốc cây là Node `=`**

## 4. Checklist khi Implement (Code)

- [ ] **Class Node:** Cần 3 thuộc tính: `value`, `left`, `right`.
- [ ] **Precedence Map:** Dictionary quy định độ ưu tiên (`{'*': 2, '+': 1...}`).
- [ ] **Xử lý số nhiều chữ số:** Bài này chỉ yêu cầu 1 ký tự, nhưng nếu mở rộng cần vòng lặp `while` để ghép chuỗi số.
- [ ] **Lỗi thường gặp:**
    - Quên `pop` dấu `(` ra khỏi stack sau khi gặp `)`.
    - Gán nhầm `left` và `right` khi dựng cây từ Stack.
    - Điều kiện vòng lặp `while` trong Infix->Postfix cần kiểm tra Stack không rỗng trước khi `peek`.

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class ExpressionTree:
    def __init__(self):
        # Độ ưu tiên toán tử
        self.precedence = {'=': 0, '+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

    # --- BƯỚC 1: CHUYỂN INFIX -> POSTFIX ---
    def infix_to_postfix(self, expression):
        stack = []
        postfix = []
        
        for char in expression:
            if char.isalnum(): # Là toán hạng (chữ hoặc số)
                postfix.append(char)
            elif char == '(':
                stack.append(char)
            elif char == ')':
                while stack and stack[-1] != '(':
                    postfix.append(stack.pop())
                stack.pop() # Bỏ dấu '('
            else: # Là toán tử
                while (stack and stack[-1] != '(' and 
                       self.precedence.get(char, 0) <= self.precedence.get(stack[-1], 0)):
                    postfix.append(stack.pop())
                stack.append(char)
        
        while stack:
            postfix.append(stack.pop())
        return "".join(postfix)

    # --- BƯỚC 2: DỰNG CÂY TỪ POSTFIX ---
    def build_tree(self, postfix_expr):
        stack = []
        
        for char in postfix_expr:
            if char.isalnum():
                node = Node(char)
                stack.append(node)
            else:
                node = Node(char)
                # Lưu ý: Pop bên phải trước, bên trái sau
                if stack: node.right = stack.pop()
                if stack: node.left = stack.pop()
                stack.append(node)
        
        return stack[0] if stack else None

    # --- BƯỚC 3: CÁC HÀM DUYỆT (TRAVERSAL) ---
    # 1. Tiền tự (Pre-order): Gốc -> Trái -> Phải (Dạng Prefix)
    def traverse_pre_order(self, node):
        if node:
            print(node.value, end=" ")
            self.traverse_pre_order(node.left)
            self.traverse_pre_order(node.right)

    # 2. Trung tự (In-order): Trái -> Gốc -> Phải (Dạng Infix gốc)
    def traverse_in_order(self, node):
        if node:
            # Thêm ngoặc nếu cần để dễ nhìn (logic nâng cao), ở đây in cơ bản
            if not node.value.isalnum(): print("(", end="") 
            self.traverse_in_order(node.left)
            print(node.value, end="") # In không có dấu cách cho giống biểu thức
            self.traverse_in_order(node.right)
            if not node.value.isalnum(): print(")", end="")

    # 3. Hậu tự (Post-order): Trái -> Phải -> Gốc (Dạng Postfix)
    def traverse_post_order(self, node):
        if node:
            self.traverse_post_order(node.left)
            self.traverse_post_order(node.right)
            print(node.value, end=" ")

# --- CHẠY THỬ ---
expr_str = "A=3+5*2-7-(6+b)"
tree_tool = ExpressionTree()

print(f"Biểu thức gốc: {expr_str}")

# 1. Chuyển đổi
postfix = tree_tool.infix_to_postfix(expr_str)
print(f"Postfix (RPN): {postfix}")

# 2. Dựng cây
root = tree_tool.build_tree(postfix)

# 3. Kết quả duyệt
print("\n--- KẾT QUẢ DUYỆT ---")
print("Tiền tự (Pre-order - Prefix): ", end="")
tree_tool.traverse_pre_order(root)
print()

print("Trung tự (In-order - Infix):  ", end="")
tree_tool.traverse_in_order(root)
print()

print("Hậu tự (Post-order - Postfix):", end="")
tree_tool.traverse_post_order(root)
print()
```
