# Cấu Trúc Dữ Liệu: Cây Tổng Quát (General Tree)

## 1. Tư duy cốt lõi: Cây là gì?

Cây tổng quát là một cấu trúc dữ liệu **phi tuyến tính** (non-linear), mô phỏng cách tổ chức phân cấp (hierarchical) trong thực tế (như gia phả dòng họ, sơ đồ công ty).

> **Định nghĩa:** Cây tổng quát là một tập hợp các nút (nodes), trong đó có một nút đặc biệt gọi là **Gốc (Root)**. Các nút còn lại được chia thành các tập hợp rời nhau, mỗi tập hợp lại là một cây con (subtree).


## 2. Các thuật ngữ chuyên môn (Vocabulary)

### Thành phần cơ bản
* **Nút (Node):** Đơn vị chứa dữ liệu (Data).
* **Cạnh (Edge):** Đường nối giữa hai nút. Nếu cây có $N$ nút, sẽ có $N-1$ cạnh.
* **Gốc (Root):** Nút duy nhất nằm trên cùng, không có cha.

### Quan hệ họ hàng
* **Cha (Parent):** Nút nằm ngay phía trên.
* **Con (Child):** Nút nằm ngay phía dưới (Một cha có thể có $0 \dots N$ con).
* **Anh em (Siblings):** Các nút có cùng một cha.
* **Lá (Leaf):** Các nút nằm dưới cùng, không có con.
* **Nút trong (Internal Node):** Các nút có ít nhất một con.

### Các chỉ số đo lường
* **Cấp/Mức (Level):** Gốc là mức 0.
* **Độ sâu (Depth):** Khoảng cách từ Gốc đến nút đó.
* **Chiều cao (Height):** Khoảng cách dài nhất từ nút đó xuống Lá xa nhất.


## 3. Biểu diễn trong bộ nhớ (Quan trọng)

Vì số lượng con của một nút là không xác định, ta không thể dùng cấu trúc cố định như cây nhị phân.

### Cách 1: Danh sách các con (List of Children)
Mỗi nút chứa dữ liệu và một danh sách (Vector/List) các con trỏ trỏ tới tất cả các con.
* *Ưu điểm:* Trực quan.
* *Nhược điểm:* Tốn bộ nhớ quản lý danh sách động.

### Cách 2: Con trái nhất - Anh em ruột phải (Left-Child, Right-Sibling)
Đây là cách tối ưu nhất để chuẩn hóa cây tổng quát. Mỗi nút chỉ cần 2 con trỏ:
1.  **Left Child:** Trỏ tới đứa con *đầu tiên* (bìa trái).
2.  **Right Sibling:** Trỏ tới người *anh em kế tiếp* (ngay bên phải).

> **Logic:** Muốn tìm con thứ 2, ta đi đến con đầu, rồi nhảy sang phải theo con trỏ Sibling.


## 4. Các thao tác duyệt cây (Traversal)

### A. Duyệt theo chiều sâu (DFS - Depth First Search)
1.  **Pre-order (Tiền thứ tự):** Gốc $\rightarrow$ Các con (Trái sang Phải).
2.  **Post-order (Hậu thứ tự):** Các con (Trái sang Phải) $\rightarrow$ Gốc.

### B. Duyệt theo chiều rộng (BFS - Breadth First Search)
Duyệt theo từng tầng (Level-order) từ trên xuống dưới, từ trái qua phải.
* *Yêu cầu:* Sử dụng Hàng đợi (Queue).


## 5. Ứng dụng thực tế

1.  **Hệ thống tập tin (File Systems):** Thư mục cha chứa thư mục con và tệp tin.
2.  **DOM (Document Object Model):** Cấu trúc phân cấp của trang HTML (`html` > `body` > `div`).
3.  **Cây cú pháp (AST):** Dùng trong trình biên dịch (Compiler) để phân tích mã nguồn.


## 6. So sánh: Cây tổng quát vs Cây nhị phân

| Đặc điểm | Cây tổng quát | Cây nhị phân |
| :--- | :--- | :--- |
| **Số con** | Không giới hạn | Tối đa 2 |
| **Cài đặt** | Phức tạp hơn | Đơn giản, chuẩn mực |
| **Mục đích** | Biểu diễn phân cấp dữ liệu | Tối ưu tìm kiếm, sắp xếp |
