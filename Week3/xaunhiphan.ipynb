# BT1:
In ra tất cả các xâu nhị phân độ dài n (n<10) nhưng KHÔNG được chứa 2 bit 1 liên tiếp

```python
def generate_valid_binary_strings(n):
    """
    Hàm chính để bắt đầu quá trình sinh các xâu nhị phân hợp lệ độ dài n.
    """
    if n <= 0:
        print("Độ dài n phải là số nguyên dương.")
        return

    print(f"--- Các xâu nhị phân hợp lệ độ dài {n} ---")
    
    # Bắt đầu quá trình backtracking với một xâu rỗng ban đầu
    _backtrack(n, "")

def _backtrack(n, current_string):
    """
    Hàm đệ quy sử dụng backtracking.
    n: độ dài mục tiêu.
    current_string: xâu đang được xây dựng.
    """
    # 1. TRƯỜNG HỢP CƠ SỞ (BASE CASE)
    # Nếu xâu hiện tại đã đủ độ dài n, nó là một giải pháp.
    if len(current_string) == n:
        print(current_string)
        return  # Kết thúc nhánh đệ quy này

    # 2. TRƯỜNG HỢP ĐỆ QUY (RECURSIVE STEP)
    
    # Lựa chọn 1: Luôn có thể thêm '0'
    _backtrack(n, current_string + '0')

    # Lựa chọn 2: Chỉ thêm '1' nếu hợp lệ
    # Điều kiện hợp lệ: xâu rỗng HOẶC ký tự cuối cùng không phải là '1'
    if len(current_string) == 0 or current_string[-1] == '0':
        _backtrack(n, current_string + '1')

# --- Thực thi cho n=4 và n=8 ---
if __name__ == "__main__":
    generate_valid_binary_strings(10)
    print("\n" + "="*30 + "\n")
    generate_valid_binary_strings(8)
```
